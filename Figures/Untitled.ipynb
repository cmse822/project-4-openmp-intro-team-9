```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("results_part1_q3_4_with_threads.csv")

# Separate data for different values of N
N_values = sorted(df['N'].unique())[:3]  # Sort and select first three unique N values

# Define a list of colors for each plot
colors = ['tab:blue', 'tab:orange', 'tab:green']

# Create a figure with 1 row and 3 columns for subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plotting each N value in a separate subplot with a unique color
for i, (N, ax) in enumerate(zip(N_values, axes)):
    subset = df[df['N'] == N]
    color = colors[i]  # Choose color from the list
    ax.plot(subset['nthread'], subset['average_elapsed_time'], marker='o', label=f'N={N}', color=color)

    ax.set_title(f'Average Elapsed Time vs nthread for N={N}')
    ax.set_xlabel('Number of threads')
    ax.set_ylabel('Average Elapsed Time (s)')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.legend()
    ax.grid(True)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV files into Pandas DataFrames
dfs = {}
for n in [10, 100, 1000, 2000]:
    file_path = f"results_N{n}.csv"  # Assuming your files are named data_n10.csv, data_n100.csv, etc.
    dfs[n] = pd.read_csv(file_path)

# Plotting
colors = ['b', 'g', 'r']  
colors = ['tab:blue', 'tab:red', 'tab:green']  

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

for i, (n, df) in enumerate(dfs.items()):
    row = i // 2
    col = i % 2
    ax = axs[row, col]

    ax.set_title(f"Elapsed Time for N = {n}")
    ax.set_xlabel("Number of Threads")
    ax.set_ylabel("Average Elapsed Time (s)")

    for rank, color in zip(df['n_processors'].unique(), colors):
        rank_df = df[df['n_processors'] == rank]
        ax.plot(rank_df['nthread'], rank_df['average_elapsed_time'], label=f'Rank= {rank}', color=color, marker='o')

    ax.legend()
    ax.grid(True)
    ax.set_xscale('log')
    ax.set_yscale('log')

plt.tight_layout()
plt.show()


```
